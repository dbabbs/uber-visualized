```python
import json
from pprint import pprint

data = json.load(open('../data/routes.geojson'))

data
```

```python
points = []
starts = []
ends = []
for line in data['features']:
    if line['geometry']['coordinates'][0] != 'e':
        start = [round(float(line['geometry']['coordinates'][0][0]), 5), round(float(line['geometry']['coordinates'][0][1]), 5)]
        end = [round(float(line['geometry']['coordinates'][len(line['geometry']['coordinates'])-1][0]), 5), round(float(line['geometry']['coordinates'][len(line['geometry']['coordinates'])-1][1]), 5)]
        points.append({
            'start': start,
            'end': end
        })
        starts.append(str(start))
        ends.append(str(end))
```

```python
start_data= {}
for s in starts:
    if s in start_data:
        start_data[s] = int(start_data[s]) + 1
    else:
        start_data[s] = 1
```

```python
end_data= {}
for e in ends:
    if e in end_data:
        end_data[e] = int(end_data[e]) + 1
    else:
        end_data[e] = 1
```

```python
def square(origin):
    longitude = float(origin.split(',')[0][1:])
    latitude = float(origin.split(',')[1][0:-1])
    result = [
        [longitude, latitude],
        [longitude, latitude + 0.00013],
        [longitude - 0.0002, latitude + 0.00013],
        [longitude -0.0002, latitude],
        [longitude, latitude]
    ]
    return result
square('[-122.33335375785828,47.61162776396617]')
```

```python
start_out = {
   "features": [],
   "type": "FeatureCollection"
}

end_out = {
   "features": [],
   "type": "FeatureCollection"
}
```

```python
counter = 0
for z in start_data:
    counter += 1
    start_out['features'].append({
          "type": "Feature",
          "properties": {
             "base_height": 0,
             "height": start_data[z] * 25,
             "color": "#FE872F"
          },
          "geometry": {
            "type": "Polygon",
            "coordinates": [square(z)]
          },
          "id": counter
       })
    
for q in end_data:
    counter += 1
    end_out['features'].append({
          "type": "Feature",
          "properties": {
             "base_height": 0,
             "height": end_data[q] * 25,
             "color": "#5EBA7D"
          },
          "geometry": {
            "type": "Polygon",
            "coordinates": [square(q)]
          },
          "id": counter
       })

import json
with open('../data/start.geojson', 'w') as fp:
    json.dump(start_out, fp)

with open('../data/end.geojson', 'w') as fp:
    json.dump(end_out, fp)
```
