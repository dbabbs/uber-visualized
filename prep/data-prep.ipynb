## Part 2: Route shapes from origin to destination

```python
def route(start,end):
    route_base = ("https://route.cit.api.here.com/routing/7.2/calculateroute.json?app_id="
    +hereId+ "&app_code=" + hereCode + "&waypoint0=geo!" + start + "&waypoint1=geo!" 
    + end + "&mode=fastest;car;traffic:disabled&routeattributes=shape")
    request = Request(route_base)
    try:
        response = urlopen(request)
        read_response = response.read()
        data = json.loads(read_response)
        shape = data['response']['route'][0]['shape']
        wrangled_shape = []
        for s in shape:
            temp = s.split(',')
            wrangled_shape.append([float(temp[1].encode('ascii')), float(temp[0].encode('ascii'))])
        distance = data['response']['route'][0]['summary']['distance']
        return [wrangled_shape, distance]
    except:
        return "error"

j = route('47.4406024,-122.2962891', '47.68545,-122.31718')
```

```python
for x in range(0,len(json_data)):
    result = route(json_data[x]['start_coordinates'], json_data[x]['end_coordinates'])
    json_data[x]['path'] = result[0]
    json_data[x]['distance'] = result[1]
```

# Part 3: Transforming to geojson & data driven styling

```python
layer = {
     "type": "FeatureCollection",
     "features": []
}
```

```python

for x in range(0,len(json_data)):
    layer['features'].append({
        "type": "Feature",
        "properties": {
            "id": json_data[x]['trip_id'],
#             "price": json_data[x]['price'].encode('ascii', 'ignore'),
            "date": json_data[x]['date'],
            "distance": json_data[x]['distance'],
#             "color": '#FFCF31',
            "city": json_data[x]['city'],
            "car_type": json_data[x]['car_type'],
            "month": json_data[x]['date'].split("/")[0]
        },
        "geometry": {
            "type": "LineString",
            "coordinates": json_data[x]['path']
        }
    })
    
##FFCF31
with open('./../data/routes.geojson', 'w') as fp:
    json.dump(layer, fp)
```
